###  **`parameters`**
A principal função da classe é realizar a seleção de colunas específicas do dataset, a fim de realizar o tratamento dos dados.

#### **Parâmetros**
* `df` - recebe um dataset

#### **Funções**
* `list_columns` - A função irá receber uma lista com os nomes das colunas que possuem dados no formato json do dataset original, e partir disso, criará novos nomes de colunas para cada informação salva em json.

* `columns_date` - Faz a seleção das colunas que contém apenas dados no formato `datetime` (possui um parâmetro chamado `origin`, que caso não seja preenchido, a função interpretará que o a base já foi tratada, caso contrário é realizado tratamento dos dados para depois realizar a seleção).

* `slice_columns` - selecionar colunas específicas do dataset. 
---

```
# Trabalha com a seleção dos nomes das colunas
class parameters_pandas:
    def __init__(self, df):
        self.columns_df = df.columns
        self.word1 = 'data_inicio_' 
        self.word2 = 'data_fim_'
        self.word3 = 'diferenca_em_minutos_'
        self.list_words = [self.word1, self.word2, self.word3]
    
    #Responsável por criar o nome das novas colunas criadas pela classe data_clear
    def list_columns(self, start=13, stop=None):
        name_columns = self.columns_df[start:stop]
        return_list = []
        for c in name_columns:
            for i in range(3):
                word = f'{self.list_words[i]}{c}'
                return_list.append(word)
        return return_list
    # Extraí o nome das colunas que possuiem dados do tipo datetime
    def columns_date(self, start=2, stop=None, step=3, origin=None):
        list_col = list(self.columns_df.copy())
        if origin == None:
              list_col = self.list_columns()
        for c in list_col[start:stop:step]:
              list_col.remove(c)
        return list_col

    # Fatia a lista de colunas
    def slice_columns(self, start=0, stop=None, step=None):
        return self.columns_df[start:stop:None]
```

### **`data_clear`**
A classe recebe o dataset original, com colunas que possuem dados em formato json. É realizado o tratamento da base, retornando ao final um dataset com o **json** em colunas separadas.

#### **Parâmetros**
* `df` - dataset original sem nenhum tratamento 

#### **Funções**
* `extract_and_create` - responsável por extrair os dados em **json** e depositá-los em novas colunas.
---

```
import json

#Extrai os dados do dataset original
class data_clear_pandas:
    def __init__(self, df):
        self.df = df
        self.parameters = parameters_pandas(df)
    
    # Extraí os dados em json, criando novas colunas
    def extract_and_create(self, transform=None):
        l = []
        self.return_list = []
        interlist = []
        all_df_list = []
        for i in self.parameters.slice_columns(13):
            time_list = self.df[i].iloc[:]
            for j in time_list:
                objeto = {'data_inicio': j[0], 'data_fim': j[1], 'diferenca_em_minutos': j[2]}
                interlist.append(objeto)
            self.return_list.append(interlist)
            interlist = []
        if transform == True:
            self.return_list, contador_df = self.extract_and_transform(self.return_list)
        df_lists = []
        new_df = pd.DataFrame()
        for c in self.return_list:
            mini_df = pd.DataFrame(c)
            all_df_list.append(mini_df)
        new_df = pd.concat(all_df_list, axis=1)
        new_df.columns = self.parameters.list_columns()
        #     columns_date = self.parameters.columns_date()
        #     for i in columns_date:
        #       new_df[i] = pd.to_datetime(new_df[i].astype(str), errors='coerce', utc=True)
        df_tratado = pd.concat([self.df[self.df.columns[:13]], new_df], axis=1)
        #     if transform == True:
        #       df_cont = pd.DataFrame(contador_df)
        #       df_tratado = [df_tratado, df_cont]
        return df_tratado

    def extract_and_transform(self, return_list):
        return_list = return_list
        return_list_contadores = []
        c = 0
        contNan = 0
        contOk = 0
        contNull = 0
        for i in range(len(return_list)):
            lista_interna = return_list[i]
            for j in range(len(lista_interna)):
                if list(lista_interna[j].values()).count(None) == 3:
                    return_list[i][j] = {'data_inicio': 0, 'data_fim': 0,'diferenca_em_minutos': 0}
                    contNull += 1
                elif list(lista_interna[j].values()).count(None) == 2 or list(lista_interna[j].values()).count(None) == 1:
                    return_list[i][j] = {'data_inicio': np.nan, 'data_fim': np.nan,'diferenca_em_minutos': np.nan}
                    contNan += 1
                else:
                    contOk += 1
            return_list_contadores.append({'incompleto': contNan, 'completo': contOk, 'vazio': contNull})
            contNan = 0
            contOk = 0
            contNull = 0
        return return_list, return_list_contadores
```

### **`corr_analysis`**
Sendo uma classe filha da `data_clear`, esta possui certas funcionalidades parecidas. No entanto, se diferencia da anterior pois trabalha as correlações que existem em um dataframes.

#### **Parâmetros**
* `df` - recebe um dataset

#### **Funções**

* `analysis` - Responsável por identificar as correlações do dataframe, se diferencia da função nativa do pandas, já que esta busca a média aritmética do início e fim de um processo (descarta a diferença de minutos).

* `create_reports_corr` - Cria um texto com os níveis de correlação das colunas (útil para a documentação da matriz).
---

```
# Analisa a correlação que existe no dataset
class corr_analysis_pandas(data_clear_pandas):
    def __init__(self, df):
        super().__init__(df)
        self.columns_df_origin = df.columns
        self.df_corr_analysis = df
        self.corr_analysis = self.extract_and_create()
        self.columns_date = parameters_pandas(self.corr_analysis).columns_date(start=14, origin=False)[13:]
        self.corr = self.corr_analysis[self.columns_date].isnull().corr()
        self.corrm = self.corr[(self.corr.select_dtypes(include='number') >= 0.45) & (self.corr.select_dtypes(include='number') <= 1.00)]
        self.corrm.fillna(0, inplace=True)
    # Analisa a correlação das colunas retornando um dataset
    def analysis(self):
        self.corr = self.corr_analysis.isnull().corr()
        lista = []
        self.corrm.loc[self.corrm.columns[0]]
        for i in range(len(self.corrm.loc[self.corrm.columns])):
            lista.append(list(self.corrm.loc[self.corrm.columns[i]]))
        cont = soma = media = conta = 0
        medialist = []
        listaMaiorMedia = []
        for i in lista[::2]:
            cont = 0
            for j in i:
                soma += j
                if cont == 1:
                    media = soma / 2
                    soma = 0
                    medialist.append(media)
                cont += 1
                if cont == 2:
                    cont = 0
            listaMaiorMedia.append(medialist)
            medialist = []
        media_df = pd.DataFrame(listaMaiorMedia, columns=self.corrm.columns[0::2])
        media_df.columns = self.columns_df_origin[13:]
        media_df.index = self.columns_df_origin[13:]
        return media_df

    # Cria um texto a partir da correlação do dataset criado na função analysis
    def create_reports_corr(self):
        media_df = self.analysis()
        print("## **Nível de correlação entre as colunas**")
        print('-------')
        for i in media_df.index:
            print(f'Coluna `{i}`:')
            for j in media_df.index:
                if media_df[j].loc[i] > 0:
                    print(f'* `{j}`: {(media_df[j].loc[i]):.2f}')
            print('-------')
```

```
#Função para verificar valores nulos nas filiais
def values_null_in_filiais(df, lista=None, date=None):
    tamanho = 0
    list_filiais = lista
    if type(lista) != list:
        list_filiais = list(df['nome_filial'].unique())
    if date != None:
        list_filiais = df[df['nome_filial'] > date]['nome_filial'].unique()
    total_df = {}
    dic = {}
    for i in list_filiais:
        if date != None:
            total_df[i] = df[(df['nome_filial'] > date) & (df['nome_filial'] == i)]
        else:
            total_df[i] = df[df['nome_filial'] == i]
        dic[i] = [total_df[i][total_df[i]['numero_pedido'].isnull() == True].shape[0],  # total values null
                  total_df[i][total_df[i]['numero_pedido'].isnull() == False].shape[0], # total values 'pedidos'
                  total_df[i].shape[0],                                      # size df 'filial'
                  total_df[i][total_df[i]['id_status_final'] == 6].shape[0], # total status 6
                  total_df[i][total_df[i]['id_status_final'] == 8].shape[0], # total status 8
                  total_df[i][total_df[i]['id_status_final'] == 0].shape[0], # total status 0
                  list(total_df[i][total_df[i]['id_status_final'] == 6]['data_fim_entrega_finalizada'].tail(1))
                 ]
        
    return total_df, dic
```
